```python
# Clean the merged dataset
merged_data['genres'] = merged_data['genres'].fillna('Unknown')
merged_data['release_date'] = pd.to_datetime(merged_data['release_date'])
merged_data['release_year'] = merged_data['release_date'].dt.year
```
